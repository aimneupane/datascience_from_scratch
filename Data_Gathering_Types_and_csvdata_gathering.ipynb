2.How to gather datasets from csv file format using url?
    * We have already imported the libraries but, while dealing with Url we need to import requests because we need to use get    method to exract the data from the internet archive. Also we have to import StringIO libraries 
* import requests
*  From io import StringIO
*   url= ‘’
*   When using the get method (or other HTTP methods like post, put, etc.), you might need to include specific headers to    *interact with web services correctly.
* headers={‘’}
* if you don’t know how to use headers then go to whatismybrowser.com, then go to settings and get my agent.
* Then, we have to extract using get where requests .get attributes take two parameters one is url and another one is headers.
* req=requests.get(url,headers)
* response=StringIO(req.text)
* df=pd.Dataframe(response)
* Then put the response object in Dataframe in order to create a tabular data.

```python
# through url for csv:
import requests
from io import StringIO
```

```python
url="https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv"
headers={"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req=requests.get(url,headers=headers)
data=StringIO(req.text)
df=pd.read_csv(data)
df
```
